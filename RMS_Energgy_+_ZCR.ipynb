```python
 import librosa
 import librosa.display
 import IPython.display as ipd
 import matplotlib.pyplot as plt
 import numpy as np
```

```python
# load audio files
file_1 = "/content/drive/MyDrive/Data/genres_original/blues/blues.00000.wav"
file_2 = "/content/drive/MyDrive/Data/genres_original/disco/disco.00000.wav"
file_3 = "/content/drive/MyDrive/Data/genres_original/pop/pop.00002.wav"

blues, _ = librosa.load(file_1)
disco, _ = librosa.load(file_2)
pop, _ = librosa.load(file_3)
```

```python
# extract RMSE with librosa

FRAME_LENGTH = 1024
HOP_LENGTH = 512

rms_blues = librosa.feature.rms(blues, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)[0]
rms_disco = librosa.feature.rms(disco, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)[0]
rms_pop = librosa.feature.rms(pop, frame_length=FRAME_LENGTH, hop_length=HOP_LENGTH)[0]
```

```python
rms_blues.shape
```

```python
# plot the RMSE for all the music pieces

frames1 = range(0, rms_blues.size)
t1 = librosa.frames_to_time(frames1, hop_length=HOP_LENGTH)

frames2 = range(0, rms_disco.size)
t2 = librosa.frames_to_time(frames2, hop_length=HOP_LENGTH)

frames3 = range(0, rms_pop.size)
t3 = librosa.frames_to_time(frames3, hop_length=HOP_LENGTH)

plt.figure(figsize=(15, 17))

plt.subplot(3, 1, 1)
librosa.display.waveplot(blues, alpha=0.5)
plt.plot(t1, rms_blues, color="r")
plt.title("Blues")
plt.ylim((-1, 1))

plt.subplot(3, 1, 2)
librosa.display.waveplot(disco, alpha=0.5)
plt.plot(t2, rms_disco, color="r")
plt.title("Disco")
plt.ylim((-1, 1))

plt.subplot(3, 1, 3)
librosa.display.waveplot(pop, alpha=0.5)
plt.plot(t3, rms_pop, color="r")
plt.title("Pop")
plt.ylim((-1, 1))

plt.show()
```
